# Stock Overview

Example notebook: fetch historical prices and plot a simple chart.

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
TICKER = "AAPL"
```

```python
data = yf.download(TICKER, period="1y", progress=False)
data.head()
```

```python
close = data["Close"]
if isinstance(close, pd.DataFrame):
    close = close.iloc[:, 0]

fig, ax = plt.subplots(figsize=(10, 4))
close.plot(ax=ax, title=f"{TICKER} Close Price (1Y)")
ax.set_ylabel("Price (USD)")
plt.tight_layout()
plt.show()
```
