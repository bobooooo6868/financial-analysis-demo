> **定位：** 单股 `yfinance` 示例 + **Plotly 折线图 PNG**（`images/aapl_plotly_cover.png`，可作 Fiverr Gig 封面）。  
> **正式报告** 见 [`main.ipynb`](main.ipynb)。

本 Notebook 含 matplotlib 与 Plotly 两种可视化；Plotly 单元格导出高清 PNG。

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
TICKER = "AAPL"
```

```python
data = yf.download(TICKER, period="1y", progress=False)
data.head()
```

```python
close = data["Close"]
if isinstance(close, pd.DataFrame):
    close = close.iloc[:, 0]

fig, ax = plt.subplots(figsize=(10, 4))
close.plot(ax=ax, title=f"{TICKER} Close Price (1Y)")
ax.set_ylabel("Price (USD)")
plt.tight_layout()
plt.show()
```

## Plotly line chart → PNG (Fiverr / portfolio cover)

Interactive line chart exported to `images/aapl_plotly_cover.png` (1280×720, 2× scale for crisp thumbnails).

```python
from pathlib import Path

import plotly.graph_objects as go

ROOT = Path.cwd().parent if (Path.cwd().parent / "images").exists() else Path.cwd()
OUT = ROOT / "images" / "aapl_plotly_cover.png"
OUT.parent.mkdir(parents=True, exist_ok=True)

close_series = close.iloc[:, 0] if isinstance(close, pd.DataFrame) else close

fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=close_series.index,
        y=close_series.values,
        mode="lines",
        name=TICKER,
        line=dict(color="#2563eb", width=2.5),
    )
)
fig.update_layout(
    title=dict(text=f"{TICKER} Close Price (1Y)", x=0.5, xanchor="center"),
    xaxis_title="Date",
    yaxis_title="Price (USD)",
    template="plotly_white",
    hovermode="x unified",
    width=1280,
    height=720,
    margin=dict(l=60, r=40, t=80, b=60),
)

fig.show()
fig.write_image(OUT, width=1280, height=720, scale=2)
print(f"Saved cover PNG: {OUT}")
```
